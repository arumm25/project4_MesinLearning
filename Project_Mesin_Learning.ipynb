```python
import matplotlib.pyplot as plt
import seaborn as sns

# Asumsikan df sudah dibaca sebelumnya dan trend_yearly sudah dibuat
# (kalau belum, jalankan bagian ini dulu)
trend_yearly = df.groupby('Year')[['Movies Released', 'Gross', 'Tickets Sold']].sum().reset_index()
trend_yearly['Net Income'] = trend_yearly['Gross'] * 0.4

# Filter tahun 1995–2015
net_income_1995_2015 = trend_yearly[(trend_yearly['Year'] >= 1995) & (trend_yearly['Year'] <= 2015)]

# Buat grafik garis
plt.figure(figsize=(12, 6))
sns.lineplot(data=net_income_1995_2015, x='Year', y='Net Income', marker='o', color='purple')
plt.title('Pendapatan Bersih Film per Tahun (1995–2015)', fontsize=14)
plt.ylabel('Pendapatan Bersih (USD)', fontsize=12)
plt.xlabel('Tahun', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Normalisasi nilai agar setara skala (persentase terhadap nilai maksimum)
data_norm = net_income_1995_2015.copy()
data_norm['Net Income (%)'] = data_norm['Net Income'] / data_norm['Net Income'].max() * 100
data_norm['Gross (%)'] = data_norm['Gross'] / data_norm['Gross'].max() * 100
data_norm['Movies Released (%)'] = data_norm['Movies Released'] / data_norm['Movies Released'].max() * 100

# Plot gabungan
plt.figure(figsize=(14, 6))
sns.lineplot(data=data_norm, x='Year', y='Net Income (%)', label='Pendapatan Bersih', marker='o', color='purple')
sns.lineplot(data=data_norm, x='Year', y='Gross (%)', label='Pendapatan Kotor', marker='o', color='green')
sns.lineplot(data=data_norm, x='Year', y='Movies Released (%)', label='Jumlah Film Dirilis', marker='o', color='blue')

plt.title('Perbandingan Kenaikan: Pendapatan Bersih, Kotor, dan Jumlah Film (1995–2015)', fontsize=14)
plt.ylabel('Skala Normalisasi (%)')
plt.xlabel('Tahun')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, f1_score

# Assuming x and y are already defined (replace with your actual feature and target variables)
# If you haven't defined x and y, please do so first.
# For example:
df = pd.read_csv('ThrowbackDataThursday Week 11 - Film Genre Stats.csv')

x = df[['Movies Released', 'Gross', 'Tickets Sold']]
y = df['Genre'] # Replace 'Genre' with your actual target column

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y_encoded, test_size=0.2, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
print("Akurasi:", accuracy_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred, average='macro'))
print("Laporan Klasifikasi:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5),
    "Support Vector Machine": SVC(kernel='linear'),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(random_state=42)
}
```

```python
results = {}
```

```python
# Latih dan evaluasi masing-masing model
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='macro')
    results[name] = {"Akurasi": acc, "F1 Score": f1}
```

```python
# Tampilkan hasil perbandingan
for model_name, scores in results.items():
    print(f"{model_name} - Akurasi: {scores['Akurasi']:.4f} | F1 Score: {scores['F1 Score']:.4f}")
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

# Inisialisasi model
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Support Vector Machine": SVC(kernel='linear'),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(random_state=42)
}

# Menyimpan hasil rata-rata skor
cv_results = {}

for name, model in models.items():
    acc_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='accuracy')
    f1_scores = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='f1_macro')

    cv_results[name] = {
        "Akurasi Mean": np.mean(acc_scores),
        "Akurasi Std": np.std(acc_scores),
        "F1 Mean": np.mean(f1_scores),
        "F1 Std": np.std(f1_scores)
    }

# Cetak hasilnya
for name, scores in cv_results.items():
    print(f"{name} - Akurasi: {scores['Akurasi Mean']:.4f} ± {scores['Akurasi Std']:.4f} | "
          f"F1 Score: {scores['F1 Mean']:.4f} ± {scores['F1 Std']:.4f}")
```
